# **Load and Combine Datasets**

```python
# Load and combine datasets for training
training_data = []
# Load and combine datasets for training
training_data = []
for name, path in dataset_paths.items():
    df = pd.read_csv(path)
    df['Dataset'] = 'Prediction'  # Add a column to track the source dataset
    training_data.append(df)

combined_data = pd.concat(training_data, ignore_index=True)
```

# **Define X and Y**

```python
X = combined_data.drop(['Fantasy Points', 'Dataset'], axis=1)
y = (combined_data['Fantasy Points'] >= 20).astype(int)
```

# **Split Train and Test Data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=100)
```

```python
X_test
```

# **Scale the features**

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# **Initialize the RandomForestClassifier**

```python
model = RandomForestClassifier(n_estimators=7, random_state=100)
```

# **Fit Model the training data**

```python
model.fit(X_train, y_train)
```

# **Predictions on test set**

```python
y_pred = model.predict(X_test)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Function to select the dataset based on the NFL team name
def select_team_dataset():
    print("\nWelcome to the Gridiron Genius!")
    print("\nAvailable NFL teams for prediction:")
    for team in dataset_paths.keys():
        print(f"- {team}")

    # Prompt user to input an NFL team name and normalize the input
    team_name = input("\nEnter the NFL team name to select a dataset: ").strip().lower()

    # Validate team selection, checking for case-insensitive match
    while team_name not in [team.lower() for team in dataset_paths.keys()]:
        print("Invalid team name. Please try again.")
        team_name = input("Enter the NFL team name to select a dataset: ").strip().lower()

    # Find the actual team name (case-sensitive) from the input
    selected_team = next(team for team in dataset_paths.keys() if team.lower() == team_name)

    # Load the selected dataset dynamically based on the team name
    selected_path = dataset_paths[selected_team]
    df_player_data = pd.read_csv('https://raw.githubusercontent.com/SIMBL742/GridironGenius/refs/heads/main/dataset_2024-2025_QBs.csv')
    return df_player_data

# Function to predict a specific player's fantasy performance
def predict_player_performance(df_player_data):
    # Prompt user to input player name
    player_first_name = input("\nEnter the player's first name: ")
    player_last_name = input("Enter the player's last name: ")

    # Check if player exists in the dataset
    player_info = df_player_data[
        (df_player_data['First Name'] == player_first_name) &
        (df_player_data['Last Name'] == player_last_name)
    ]

    if player_info.empty:
        print(f"\nNo data found for {player_first_name} {player_last_name}. Please check the spelling or ensure the player has sufficient NFL stats.")
        return

    # Process player data for prediction
    input_data = player_info[X.columns].values
    input_data_df = pd.DataFrame(input_data, columns=X.columns)  # Ensure proper column names

    # Extract the number of games the player has played
    num_games_played = player_info['Games Played'].values[0]

    # Scale the input data
    input_scaled = scaler.transform(input_data_df)

    # Make a prediction using the trained model
    prediction = model.predict(input_scaled)
    prediction_label = "BOOM" if prediction[0] == 1 else "BUST"


    # Get feature importances from the model
    feature_importances = model.feature_importances_

    # Create a DataFrame to display feature importances
    feature_importance_df = pd.DataFrame({
        'Feature': X.columns,
        'Importance': feature_importances
    })


    # Display prediction result with the number of games played
    print(f"\nGridiron Genius predicts {player_first_name} {player_last_name} will be a {prediction_label} after using data from the last {num_games_played} games.")
    # Sort by importance in descending order
    feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
    print("\nTop contributing factors for this prediction:")
    feature_importance_df['Importance'] = (feature_importance_df['Importance'] * 100).map('{:.2f}%'.format)
    print(feature_importance_df.head(7).to_string(index=False))  # Display top 3 features



# Call the functions to execute the workflow
df_player_data = select_team_dataset()
predict_player_performance(df_player_data)
```
