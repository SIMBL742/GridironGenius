# **Load and Combine Datasets**

```python
def select_team_dataset():
    print("\nWelcome to the Gridiron Genius!")
    print("\nAvailable NFL teams for prediction:")

    for team in team_to_df.keys():
        print(f"- {team}")

        # Ask for user input and clean it up
    team_input = input("\nEnter the NFL team name to select a dataset: ").strip().lower()

    # Keep asking until a valid team is entered
    while team_input not in [team.lower() for team in team_to_df.keys()]:
        print(" Invalid team name. Please try again.")
        team_input = input("Enter the NFL team name to select a dataset: ").strip().lower()

    selected_team = next(team for team in team_to_df if team.lower() == team_input)

    df = team_to_df[selected_team]

    return df

selected_team_df = select_team_dataset()

```

# **Define X and Y**

```python
X = selected_team_df.drop(['Fantasy Points'], axis=1)
y = (selected_team_df['Fantasy Points'] >= 20).astype(int)
```

```python
# Compute means & std for defense scaling
def_mu = X.mean()
def_sigma = X.std()

# Apply Z-score scaling to defense features
X_scale = (X - def_mu) / def_sigma

X_scale
```

# **Split Train and Test Data**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=.2, random_state=100)
```

# **Load Player Data**

```python
df_player_data = pd.read_csv('https://raw.githubusercontent.com/SIMBL742/GridironGenius/refs/heads/main/dataset_2024-2025_QBs.csv')
df_player_data
```

# **Scale QB Features**


```python
# Independent scaling for player data
qb_features = X.columns
qb_mu = df_player_data[qb_features].mean()
qb_sigma = df_player_data[qb_features].std()

df_player_data_z = df_player_data.copy()
for feature in qb_features:
    df_player_data_z[feature] = (df_player_data[feature] - qb_mu[feature]) / qb_sigma[feature]
```

```python
df_player_data_z
```

# **Predict Fantasy Player**

```python


def predict_player_performance(df_player_data_z):
    player_first_name = input("\nEnter the player's first name: ")
    player_last_name = input("Enter the player's last name: ")

    player_info = df_player_data_z[
        (df_player_data_z['First Name'] == player_first_name) &
        (df_player_data_z['Last Name'] == player_last_name)
    ]

    if player_info.empty:
        print(f"\nNo data found for {player_first_name} {player_last_name}. Please check the spelling.")
        return

    # Grab scaled features (already scaled!)
    input_data = player_info[X.columns]

    # Predictions
    nn_prediction = nn_model.predict(input_data)

    # Display Results
     # Display Results
    print(f"Neural Network predicts: {'BOOM' if nn_prediction[0]==1 else 'BUST'}")

predict_player_performance(df_player_data_z)
```
